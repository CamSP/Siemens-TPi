# File type preprocessing

## Merge month data

```python
# List of files
files = ["AGO2020", 
         "SEP2020", 
         "OCT2020", 
         "NOV2020", 
         "DIC2020", 
         "ENE2021", 
         "FEB2021",
         "MAR2021"]
```

```python
# File merge function
def merge_data(files):
    df = pd.DataFrame()
    for file in files:
        data = pd.read_csv("../data/original/"+file+".csv", sep=";")
        df = pd.concat([df, data])
    df = df.fillna(method="ffill")
    return df.dropna()
```

```python
min_data = merge_data(files)
min_data.to_csv("../data/merged/min_data.csv")
min_data
```
