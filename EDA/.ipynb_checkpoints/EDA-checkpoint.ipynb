```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
files = ["AGO2020", 
         "SEP2020", 
         "OCT2020", 
         "NOV2020", 
         "DIC2020", 
         "ENE2021", 
         "FEB2021",
         "MAR2021"]
```

```python
def merge_data(files):
    df = pd.DataFrame()
    for file in files:
        data = pd.read_csv("./"+file+".csv", sep=";")
        df = pd.concat([df, data])
    df = df.fillna(method="ffill")
    return df.dropna()
```

```python
data = merge_data(files)
data
```

```python
data.to_csv("data.csv")
```
